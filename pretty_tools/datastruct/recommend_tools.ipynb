```python
# 统计相关性的函数 scipy.stats.pearsonr

from itertools import combinations

import numpy as np
from scipy import stats

import networkx as nx

import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
COLORS = [
    '#00B0F0',
    '#FF0000'
]
# Let's set random seed for reproducibility
np.random.seed(45)

# `b` represents our confounder
b = np.random.rand(100)

# `a` and `c` are causally independent of each other, but they are both children of `b`
a = b + .1 * np.random.rand(100)
c = b + .3 * np.random.rand(100)

# Let's check correlation between `a` and `c`
coef, p_val = stats.pearsonr(a, c)

print(coef)

variables = {
    'a': a,
    'b': b,
    'c': c
}

plt.figure(figsize=(12, 7))

for i, (var_1, var_2) in enumerate([('b', 'a'), ('b', 'c'), ('c', 'a')]):

    color = COLORS[1]

    if 'b' in [var_1, var_2]:
        color = COLORS[0]

    plt.subplot(2, 2, i + 1)
    plt.scatter(variables[var_1], variables[var_2], alpha=.8, color=color)

    plt.xlabel(f'${var_1}$', fontsize=16)
    plt.ylabel(f'${var_2}$', fontsize=16)

plt.suptitle('Pairwise relationships between $a$, $b$ and $c$')
plt.subplots_adjust(hspace=.25, wspace=.25)
plt.show()
```

```python
#! 用 scipy 生成某种分布的数据
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

sample_size = 1000

u_0 = stats.uniform() #* 一个代表均匀分布的随机变量生成器
u_1 = stats.norm() #* 一个代表正态（高斯）分布的随机变量生成器
# rvs (random variates)
data_0 = u_0.rvs(sample_size)
data_1 = u_1.rvs(sample_size)
size_t = 2
plt.figure(figsize=(8, 8), dpi=120)
# -----------------
plt.subplot(2, 2, 1)
plt.scatter(np.arange(sample_size), data_0, s=size_t)

plt.subplot(2, 2, 2)
plt.scatter(np.arange(sample_size), data_1, s=size_t)

plt.subplot(2, 2, 3)
plt.scatter(np.arange(sample_size), np.sort(data_0), s=size_t)

plt.subplot(2, 2, 4)
plt.scatter(np.arange(sample_size), np.sort(data_1), s=size_t)


plt.show()
```
